### inwhich year there was the highest average voting

```python
data.groupby('release_year')['vote_average'].mean()
```

```python
labels=['2014','2015']
values=[5.986792,5.878010]
plt.bar(labels,values)
plt.title('votes by year')
plt.xlabel('year')
plt.ylabel('vote')
plt.show()
```

### inwhich year there was the highest average revenue

```python
data.groupby('release_year')['revenue'].mean()
```

```python
labels=['2014','2015']
values=[6.525108e+07,3.909238e+07]
plt.bar(labels,values)
plt.title('revenue by year')
plt.xlabel('year')
plt.ylabel('revenue')
plt.show()
```

### Display Top 10 Lengthy Movies Title and Runtime 

```python
data.columns
```

```python
Top10=data.nlargest(10,'runtime')[['original_title','runtime']].set_index('original_title')
Top10
```

```python
sns.barplot(x='runtime',y=Top10.index ,data=Top10)
plt.title('The top10 movies by runtime')
plt.show()
```

### Display Number of Movies Per Year

```python
data['release_year'].value_counts()
```
